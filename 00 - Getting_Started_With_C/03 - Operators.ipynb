## Operator:
    An operator is a symbol that tells the compiler to perform specific mathematical or logical operations.
## Operand:
    An operand could be any literal(constant), variable, and anything on which an operator performs an operation.
    

Example:
         
         PI * radius * radius 
         10 + 20
         sizeof(int)
         int x = 5;

PI, radius, 10, 20, sizeof, x, 5 -> Operand
      
*, +, int,=    -> Operator

## Types Of Operators
   * Arithmetic         
   * Logical            
   * Assignment         
   * Relational
   * Increment and Decrement
   * Bitwise Operator
   * Special Operator

```python
#include <stdio.h>

int main(){ 
    //Arithmetic Operators: +, -, *, /, and %

    int x = 12, y = 4, sum, sub, mul, div, mod;

    sum = x + y;
    sub = x - y;
    mul = x * y;
    div = x / y;
    mod = x % y;

    printf("Sum of x and y : %d\n", sum);
    printf("Difference of x and y : %d\n", sub);
    printf("Multipication of x and y : %d\n", mul);
    printf("division of x and y : %d\n", div);
    printf("Remainder of x and y : %d\n", mod);
    printf("Remainder of x and y : %d\n", 12 % 5);

    return 0;
}
```

### Precedence And Associativity
    The higher Precedence is for square brackets

```python
#include <stdio.h>

int main(){ 
    //Arithmetic Operators: +, -, *, /, and %
    //Precedence: {*, /, %} > {+, -}
    //Associativity: Left to Right

    int result = 3 + 1 - 7 * 5 % 3 / 2;
    // => 3 + 1 - 35 % 3 / 2
    // => 3 + 1 - 2 / 2
    // => 3 + 1 - 1
    // => 4 - 1
    // => 3
    printf("%d",result);

    return 0;
}
```

```python
#include <stdio.h>

int main(){ 
    /*Relational Operators
        - A condition in C returns either 0 or 1
            - 0 represents a false condition
            - 1 represents a true condition */
    
    int x = 2;
    int y = 3, z = 3;

    int r1 = x > y;     //False: r1 = 0
    int r2 = 3 >= 2;     //True:  r2 = 1
    
    int r3 = x < y;     //True: r3 = 1
    int r4 = y <= x;     //False:  r4 = 0
    
    int r5 = y == z;     //True: r5 = 1
    int r6 = z != x;     //True:  r6 = 1
    
    printf("The value of r1: %d\n",r1);
    printf("The value of r2: %d\n",r2);
    printf("The value of r3: %d\n",r3);
    printf("The value of r4: %d\n",r4);
    printf("The value of r5: %d\n",r5);
    printf("The value of r6: %d\n",r6);
    return 0;
}
```

### Logical Operator
   * Logical AND
   * Logical OR
   * Logical NOT

```python
#include <stdio.h>

int main(){ 
    //Logical Operators
    //Associativity: Left to Right

    int x = 2;
    int y = 3,z = 3;
    
    //Logical AND: &&
    int a = y > x;          //True: a = 1
    int b = z >= y;         //True: b = 1
    int r1 = a && b;
    printf("The value of r1: %d\n",r1);

    int r1 = y > x && z >= y;   //r1 = 1 && 1 = 1
    printf("The value of r1: %d\n", r1);

    //Logical OR: ||
    int r2 = x > y || y > z || y != x       //r2 = 0 || 0 || 1 || 1

    //Logical NOT: !
    int r3 = !(x > y);              //r3 = !(0) = 1    

    return 0;
}
```

## Increment and Decrement Operator
   * Postfix 
       * myVariable ++
       * myVariable --
   * Prefix
       * ++myVariable
       * --myVariable
       

### Increment Operator
   * POSTFIX
       * First assign (use) the value and then increment.
   * PREFIX
       * First increment and then assign (use) the value.
       
### Decrement Operator
   * POSTFIX
       * First assign (use) the value and then decrement.
   * PREFIX
       * First decrement and then assign (use) the value.
      

```python
#include <stdio.h>

int main(){ 
    //Increment and Decrement Operator
    //Associativity: Left to Right 

    //Postfix: First assign the value and then increment/decrement 
    //Prefix: Prefix: First increment/decerement and then use the value

    int age = 10;
    int newAge = age++;

    //newAge = 10
    //age = age + 1     //age = 11

    printf("%d\n", age);
    printf("%d\n", newAge);

    int age1 = 10;
    printf("%d\n", age1++);      //age1 = 10
    //age1 = 11
    printf("%d\n", age1);        //age1 = 11
    //age1 = 11
    printf("%d\n", ++age1);        //age1 = 12
    //age1 = 12

    printf("%d\n", age1++ + ++age1);   //12 + 14 
    //age1 = 14

    printf("%d\n", age1-- + --age1);   //14 + 12 
    //age1 = 12

    return 0;
}
```
