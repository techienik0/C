# Memory Management:
   * Static Memory Allocation
           * Allocation done at compile time. ( before program execution ).
   * Dynamic Memory Allocation
           * Allocation done at runtime. ( during program execution ).

```python
#include <stdio.h>

int main(){ 

    //Static Memory Management
    //Memory is allocated before program execution.

    int x = 24;         //4 Bytes
    float scores[100];      //(100 * 8) Bytes = 800 Bytes
    char letters[10] = {'l', 'e', 'i', 'o'};       // (10 * 1)Bytes = 10 Bytes

    //It is having more disadvantages
    return 0;
}
```

## Four Library Functions of 'stdlib.h'
   * malloc()
       * Memory Allocation
   * free()
       * De-allocate Memory
   * calloc()
       * Contiguous Allocation
   * realloc()
       * Re-allocation

```python
#include <stdio.h>

int main(){ 

    // WAP to store 'n' natural numbers in an array.
    int numbers[50], n;

    puts("Enter n : ");
    scanf("%d", &n);

    //Store 'n' natural numbers in the array.
    for (int i = 0; i < n; i++){
        numbers[i] = i + 1;
    }

    //Print all natural numbers stored in the array.
    printf("The values present in the array : ");
    for (int i = 0; i < n;i++){
        printf("%d",numbers[i]);
    }
    
    return 0;
}
```

```python
#include <stdio.h>
#include<stdlib.h>

int main(){ 

    // WAP to store 'n' natural numbers.
    int *ptr, n;

    puts("Enter n : ");
    scanf("%d", &n);

    // Dynamically allocate memory using malloc().
    ptr = (int *)malloc(n * sizeof(int));

    if (ptr == NULL){           //No Value Exists
        puts("Memory allocation failed");
        exit(0);
    }else{

        //Store 'n' natural numbers in the allocated memory block.
        for (int i = 0; i < n; i++){
            ptr[i] = i + 1;
        }

        //Print all natural numbers stored in the memory block.
        printf("The values present in the allocated memory block : ");
        for (int i = 0; i < n;i++){
            printf("%d",ptr[i]);
        }
    }
    return 0;
}
```

```python
#include <stdio.h>
#include<stdlib.h>

int main(){ 

    // WAP to store 'n' natural numbers.
    int *ptr, n;

    puts("Enter n : ");
    scanf("%d", &n);

    // Dynamically allocate memory using calloc().
    ptr = (int *)calloc(n , sizeof(int));

    if (ptr == NULL){           //No Value Exists
        puts("Memory allocation failed");
        exit(0);
    }else{

        //Store 'n' natural numbers in the allocated memory block.
        for (int i = 0; i < n; i++){
            ptr[i] = i + 1;
        }

        //Print all natural numbers stored in the memory block.
        printf("The values present in the allocated memory block : ");
        for (int i = 0; i < n;i++){
            printf("%d",ptr[i]);
        }
    }
    
    free(ptr);                              //Deallocate the memory
    return 0;
}
```

```python
#include <stdio.h>
#include<stdlib.h>

int main(){ 

    // WAP to store 'n' natural numbers.
    int *ptr, n;

    puts("Enter n : ");
    scanf("%d", &n);

    // Dynamically allocate memory using calloc().
    ptr = (int *)calloc(n , sizeof(int));

    if (ptr == NULL){           //No Value Exists
        puts("Memory allocation failed");
        exit(0);
    }else{

        //Store 'n' natural numbers in the allocated memory block.
        for (int i = 0; i < n; i++){
            ptr[i] = i + 1;
        }

        //Print all natural numbers stored in the memory block.
        printf("The values present in the allocated memory block : ");
        for (int i = 0; i < n;i++){
            printf("%d",ptr[i]);
        }
        int c, newSize;

        puts("\nEnter the number of more values you want to enter: ");
        scanf("%d", &c);

        newSize = c + n;

        //Dynamically re-allocate memory using realloc().
        ptr = (int *)realloc(ptr, newSize * sizeof(int));

        //Store more values.
        for (int i = n; i < newSize; i++){
            ptr[i] = i + 1;
        }

        printf("The final values stored in the allocated memory block.");
        for (int i = 0;i < newSize; i++){
            printf("%d ", ptr[i]);
        }
    }
    free(ptr);
    return 0;
}
```
